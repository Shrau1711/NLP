```python
corpus="""Cognition encompasses mental processes that acquire, store, retrieve, transform, or apply information. It is a pervasive part of mental life, helping individuals understand and interact with the world. Cognition includes perception, which organizes and interprets sensory input; memory, which stores and retrieves information; and thinking, which considers and manipulates ideas.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
documents=sent_tokenize(corpus)
```

```python
documents
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization
## Paragraphs-->words
## sentence-->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
## punctuation also gets separated
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
## full stop with not be treated as a separate word except for the last full stop in the paragraph.
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
