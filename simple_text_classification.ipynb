```python
from sklearn.datasets import fetch_20newsgroups
```

```python
## fetch categories
all_data = fetch_20newsgroups(subset='train')
print(all_data.target_names)
```

```python
## selected 4 catgories instead of all 20
categories=['alt.atheism','comp.graphics','rec.motorcycles','rec.sport.hockey']
```

```python
train_data = fetch_20newsgroups(subset='train', categories=categories)
test_data = fetch_20newsgroups(subset='test', categories=categories)
```

```python
print(train_data.data[0])
print(train_data.target[0])
print(train_data.target_names[train_data.target[0]])
```

```python
train_data = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
test_data = fetch_20newsgroups(subset='test', categories=categories, remove=('headers', 'footers', 'quotes'))
```

```python
print(train_data.data[0])
print(train_data.target[0])
print(train_data.target_names[train_data.target[0]])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train_data.data)
X_test = vectorizer.transform(test_data.data)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train, train_data.target)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
y_pred=model.predict(X_test)
accuracy=accuracy_score(test_data.target,y_pred)
report = classification_report(test_data.target, y_pred)
```
